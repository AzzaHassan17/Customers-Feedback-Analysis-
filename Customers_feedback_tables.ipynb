```python
from transformers import pipeline
```

```python
sentiment_analyzer= pipeline(task='sentiment-analysis', model='distilbert/distilbert-base-uncased-finetuned-sst-2-english', top_k=None)
sentences=['Great service!', 'I am very satisfied!', 'Not happy with the product.']
```

```python
#Analyze sentiment
results = sentiment_analyzer(sentences)

# Display the results
for text, result in zip(sentences, results):
    print(f"Text: {text}")
    # Iterate over the list of dictionaries in result
    for item in result:
        label = item['label']
        score = item['score']
        print(f"Sentiment: {label}, Score: {score:.4f}")
    print()
```

```python
# Extract the feedback sentences from the column
sentences = FactFeedback['feedback_text'].tolist()
```

```python
# Analyze sentiment
results = sentiment_analyzer(sentences)
```

```python
# Create lists to store the sentiments and scores
sentiments = []
scores = []
```

```python
# Iterate through the results and store only the first label and score
for result in results:
    # Get the first result from the list
    first_result = result[0]  # Assuming result is a list of dictionaries
    sentiments.append(first_result['label'])
    scores.append(first_result['score'])
```

```python
DimSentimentAnalysis = pd.DataFrame(columns=['Sentiment', 'Score','feedback_id'])
```

```python
# Add the results to a table in the datawarehouse
DimSentimentAnalysis['Sentiment']=sentiments
DimSentimentAnalysis['Score']=scores
DimSentimentAnalysis['feedback_id']=FactFeedback['feedback_id']
DimSentimentAnalysis.head()
```
